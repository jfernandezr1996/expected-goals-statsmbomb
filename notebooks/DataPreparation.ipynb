## `angle`

```python
df[df['angle'].isnull()][['x','y','angle','distance','player','opponentTeam','result','season',
                          'playerAssisted','lastAction','firstTime']]
```

> We observe that the x-coordinate is 120 which indicates that the ball was on the goal line and therefore when calculating the angle the calculation was not correct.

```python
a = np.sqrt((120-120)**2+(43.7-36)**2)
b = np.sqrt((120-120)**2+(43.7-44)**2)
angle = np.degrees(np.arccos(round((a**2+b**2-64)/(2*a*b))))
print("180¬∫ angle = flat angle: ", angle)

df['angle'] = df['angle'].fillna(angle)
```

## `xGoalkeeper - yGoalkeeper`

As seen in the building code of these fields, these 'NaN' values refer to those situations where the goalkeeper is not in the box surrounding the player with ball possession. In all these situations, it is understood that the goalkeeper was far from the player in possession. Accessing these values, we can see that only 1 shot was a goal, which means that these are situations with little probability of success (as indicated by the `xG` field).

Well, **how do we decide to impute these 'NaN' values on the goalkeeper's coordinates?**

Previously, we decided to analyze the relationships between the goalkeeper and the shot put coordinates.

### Shots where the player is closer to the goal than the opposing team's goalkeeper

```python
df[(df['xGoalkeeper'] < df['x']) & (df['xGoalkeeper']<100)][['x','y','xGoalkeeper','yGoalkeeper','result','xG','player',
                                   'bodyPart', 'playerTeam', 'opponentTeam', 'season', 'competition']]
```

### `NaN` values

When assigning the NaN values, we assume that it is not a shot from a player on his goal, otherwise we would understand that the goalkeeper would be close to that player, so we assume that the goalkeeper is on his goal.

```python
df[df['xGoalkeeper'].isnull()]['result'].value_counts()
```

#### Goals for which the goalkeeper's location is unknown

> We observe that the goalkeeper in this case is on the ground far from the ball and therefore her position has not been detected, which is in the center of the small area. Podemos identificar su posici√≥n como [118, 45]

#### Saved shots for which the location of the goalkeeper is unknown

> For those launches that had a `saved` result we decided to impute the missing values in the goalkeeper's coordinates using the values of `xEndLocation` and `yEndLocation`.

```python
df[df['xGoalkeeper'].isnull()][['x','y','result','xEndLocation', 'yEndLocation', 'player','bodyPart',
                                'xG','date','competition','season','playerTeam','opponentTeam']]\
                    .sort_values(by='xG', ascending=False).head()
```

#### Impute NaN values

For the rest of the cases we will use the coordinates `[118,40]`

```python
def impute_NaN_values_X(xGoalkeeper, xEndLocation, result):
    if pd.isna(xGoalkeeper):
        if result == 'Saved':
            return xEndLocation
        else: 
            return 118
    else:
        return xGoalkeeper
    
def impute_NaN_values_Y(yGoalkeeper, yEndLocation, result):
    if pd.isna(yGoalkeeper):
        if result == 'Saved':
            return yEndLocation
        else:
            if result == 'Goal':
                return 45
            else:
                return 40
    else:
        return yGoalkeeper
```

```python
df['xGoalkeeper'] = df[['result', 'xEndLocation', 'xGoalkeeper']].apply(
    lambda x: impute_NaN_values_X(x[2],x[1],x[0]), axis=1)
df['yGoalkeeper'] = df[['result', 'yEndLocation', 'yGoalkeeper']].apply(
    lambda x: impute_NaN_values_Y(x[2],x[1],x[0]), axis=1)
```

```python
print("N√∫mero de registros nulos:", df.isna().sum().sum())
```

# Feature Engineering and relationship to the target variable
---
We will carry out a treatment on some of our variables as well as the possibility of incorporating new fields. **The analysis of these variables is aimed at simplifying the levels of categories that can be taken by the different categorical variables that simplify the patterns acquired by the machine learning models that we will apply in the following section.**

We will rely on Bejamin Pavard's next goal in the World Cup for the treatment of the features.

* **Goal Benjamin Pavard against Argentina FIFA World Cup**

```python
YouTubeVideo('hK7BjNE2TTA')
```

```python
def plot_categorical_variables(df, attribute):
    if attribute == 'target':
        df_levels = pd.DataFrame(df[attribute].value_counts(normalize=True)).sort_index()
        fig = go.Figure(go.Bar(x=df_levels.index, y=df_levels[attribute], marker_color='chocolate', name='target'))
        fig.update_layout(title_text='Target Class Proportions', xaxis_title='target', yaxis_title='percentage')
    else:
        df_attribute = df[[attribute, 'target']]
        list_levels = list(df_attribute[attribute].unique())
        fig = go.Figure()
        for level in list_levels:
            level_df = pd.DataFrame(df_attribute[df_attribute[attribute]==level]['target'].value_counts()).sort_index()
            fig.add_trace(go.Bar(x=list(level_df.index), y=level_df['target'], name=level))
        fig.update_layout(xaxis_title = 'target', yaxis_title = 'count', 
                          title_text = 'Categorical Attribute (' + attribute + ') levels frecuency based on the target variable')
    fig.show()
```

```python
attributes_to_explain = ['x','y','result','situation', 'playerPosition', 'playerAssisted', 'bodyPart', 'playType', 
                          'firstTime','isCrossKeyPass', 'isCutBackKeyPass', 'xG', 'competition']
df[(df.competition=='FIFA World Cup') & (df.result == 'Goal') & (df.firstTime) & (df.player == 'Benjamin Pavard')]\
        [attributes_to_explain]
```

## `target`

We must not forget what our objective is: to predict whether a shot will be a goal or not as well as its probability of success based on a series of characteristics.

So, we must build our variable to predict, that is, the label of each example that tells us if the throw has been a goal or not. To do this, we use the attribute `result`.

```python
df['target'] = df['result'].apply(lambda x: 'isGoal' if x == 'Goal' else 'notGoal')

plot_categorical_variables(df, 'target')
```

## `firstTime`

This is a variable that indicates whether the shot was fired at the first touch as Benjamin Pavard's Goal. We show below two goals in which this field equals `True` and another one in which it equals `False`. Therefore, we decided to consider two categories for this new variable (`firstTimeCarry`): `firstTime` and `carry` indicating that it is a drive.

```python
df['firstTimeCarry'] = df['firstTime'].apply(lambda x: 'firstTime' if x else 'carry')

plot_categorical_variables(df, 'firstTimeCarry')
```

## `isCrossKeyPass, isCutBackKeyPass`

By definition of the manuals:
* `Cross Pass`: cross from the wing.
* `Cut-Back Pass`: pass back from the bye-line.

Looking at the video, we see that the shot is preceded by a center by Lucas Hernandez, which is indicated by the boolean value `True` in `isCrossKeyPass`.

We decided to create a new variable that is `previousKeyPass` with the categories: `cross`, `cutback` and `forward`, covering the different possibilities of a type of pass.

```python
def cat_previous_pass(x1,x2):
    if x1:
        return 'cut-back'
    elif x2:
        return 'cross'
    else:
        return 'forward'
    
df['previousKeyPass'] = df[['isCrossKeyPass','isCutBackKeyPass']].apply(lambda x: cat_previous_pass(x[1],x[0]), axis=1)

plot_categorical_variables(df, 'previousKeyPass')
```

## `playerPosition`

```python
new_positions = {'Goalkeeper': 'Goalkeeper', 'Left Back': 'Defender', 'Left Center Back': 'Defender', 
                 'Right Center Back': 'Defender', 'Right Back': 'Defender', 'Right Wing Back': 'Defender', 
                 'Left Wing Back': 'Defender', 'Center Back': 'Defender', 'Left Center Midfield': 'Midfielder', 
                 'Right Center Midfield': 'Midfielder', 'Center Attacking Midfield': 'Midfielder', 
                 'Left Midfield': 'Midfielder', 'Right Midfield': 'Midfielder', 'Center Defensive Midfield': 'Midfielder', 
                 'Left Defensive Midfield': 'Midfielder', 'Right Defensive Midfield': 'Midfielder', 
                 'Left Attacking Midfield': 'Midfielder', 'Right Attacking Midfield': 'Midfielder', 
                 'Center Midfield': 'Midfielder', 'Center Forward': 'Striker', 'Right Center Forward': 'Striker', 
                 'Left Center Forward': 'Striker', 'Secondary Striker': 'Striker', 'Left Wing': 'Striker', 
                 'Right Wing': 'Striker'}


df['playerPosition'] = df['playerPosition'].apply(lambda x: new_positions[x])
plot_categorical_variables(df, 'playerPosition')
```
