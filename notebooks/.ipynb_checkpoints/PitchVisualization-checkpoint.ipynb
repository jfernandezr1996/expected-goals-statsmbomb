# Representation of the shots according to their `Expected Goals` distinguishing the player and the part of the body

We will use the French national team by agreeing to the team's shooting at the World Cup.

```python
worldCup_France_Goal = df[(df['playerTeam']=='France') & (df['target']=='isGoal')]
worldCup_France_Goal = worldCup_France_Goal[['x', 'y', 'player', 'bodyPart', 'opponentTeam', 'xG']].\
                                sort_values(by='xG', ascending=False).reset_index(drop=True)
worldCup_France_Goal['player'] = worldCup_France_Goal['player'].apply(
                                        lambda x: x.split(' ')[0][0] + ". " + x.split(' ')[1] 
                                            if x.split(' ')[1] != 'Yves' else x.split(' ')[0][0] + "." + x.split(' ')[2])
worldCup_France_Goal['xG'] = round(worldCup_France_Goal['xG'], 2)
print("Número de goles de Francia en la Copa del Mundo 2018 (excluyendo penaltis):", worldCup_France_Goal.shape[0])
```

```python
draw_pitch("#195905","#faf0e6","vertical","half")

plot_shots(worldCup_France_Goal, 'x', 'y', 'xG', 'player', 'bodyPart', 'opponentTeam', 'red', 'France Goals')
plt.title('Expected Goals by France World Cup 2018')
plt.show()
```
