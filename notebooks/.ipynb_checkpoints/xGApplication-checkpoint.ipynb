## Obtain xG predictions and comparision between StatsBomb xG

```python
def obtain_xG_predictions(model, new_df, new_df_preprocess):
    proba_predict = model.predict_proba(np.array(new_df_preprocess.iloc[:,:-1]))
    xG_predictions = proba_predict[:,1]
    new_df_predictions = pd.concat([new_df.reset_index(drop=True), 
                                    pd.DataFrame(xG_predictions, columns=['xG_predict']).reset_index(drop=True)], axis=1)
    # Ajuste xG y xG_predict (R2)
    lr = LinearRegression()
    lr.fit(np.array(new_df_predictions['xG_predict']).reshape(-1,1), new_df_predictions['xG'])
    r2 = lr.score(np.array(new_df_predictions['xG_predict']).reshape(-1,1), new_df_predictions['xG'])
    lr_predict = lr.predict(np.vstack(new_df_predictions['xG_predict']))
    
    fig = px.scatter(new_df_predictions, x='xG_predict', y='xG', color='target')
    fig.add_traces(go.Scatter(x = new_df_predictions['xG_predict'], y = lr_predict.tolist(), name='regression line', 
                              line = dict(color='darkgreen')))
    fig.update_layout(title = 'Comparison between xG statsbomb and the xG predicted by the model. R2 = ' + str(round(r2, 2)))
    fig.show()
    return new_df_predictions


df_predictions = obtain_xG_predictions(logit_model_ohe, new_df, new_df_preprocess_ohe)
```

# Player Performance

we can observe those players with a higher performance than our model predicts indicating that the goals that these players got were with a very low % of success.

```python
player_performance = df_predictions.copy()
player_performance['target'] = player_performance['target'].apply(
        lambda x: 1 if x == 'isGoal' else 0)
player_performance = player_performance.\
        groupby(['player','playerTeam'])[['xG_predict','target']].sum()
player_performance = player_performance.\
    reset_index(['player','playerTeam']).reset_index(drop=True)
player_performance = player_performance.rename(
        columns={'xG_predict': 'Predicted xG sum', 
                 'target': 'Goals Scored'})
fig = px.scatter(player_performance, 
                 x = 'Predicted xG sum', y = 'Goals Scored', 
                 color = 'playerTeam', hover_data = ['player'])
fig.add_shape(type='line', 
              x0 = 0, y0 = 0, 
              x1 = round(player_performance['Goals Scored'].max()), 
              y1 = round(player_performance['Goals Scored'].max()),
              line = dict(color='MediumPurple', width=2, dash='dot'))
fig.update_layout(title_text = 'Player Performance xG', 
                  xaxis_title = 'Predicted xG sum', 
                  yaxis_title = 'Goals Scored', 
                  xaxis=dict(range = [0, round(player_performance['Predicted xG sum'].max())]))
fig.show()
```

### Result Simulation calculating the xG in each match

We will offer three results:

- The result of adding up the xG of StatsBomb predicted for each of the team's shots.
- Similarly, considering the xG predicted by our model.
- We will simulate 10K times each shot and again, we will add up the shots of each team.

```python
FranceCroatia['sim_xG_predict'] = FranceCroatia['xG_predict'].apply(
    lambda x: sum(binom.rvs(n=1, p=x, size=10000))/10000)

sim = FranceCroatia.groupby(['playerTeam'])['xG','xG_predict','sim_xG_predict'].sum()
sim.index.name = 'team'

f=pd.concat([sim, n_shots], axis=1)
f = f.rename(columns={'xG':'xG_statsBomb'})
f[['shots','xG_statsBomb','xG_predict','sim_xG_predict']]
```

### 10K simulations

```python
shots_match = FranceCroatia[['playerTeam','xG']].sort_values(by='playerTeam').reset_index(drop=True)

n_simulations = 10000
total_simulations = []
for j in range(n_simulations):
    match_result = dict()
    for team in list(shots_match.playerTeam.unique()):
        shots_team = shots_match[shots_match.playerTeam == team]
        result_shot = []
        for i in range(shots_team.shape[0]):
            result_shot.append(binom.rvs(n = 1, p = shots_team.iloc[i,1], size = 1)[0])
        match_result[team] = sum(result_shot)
    total_simulations.append(match_result)
```

* **Result Simulation**

```python
simulation_matches = pd.DataFrame(total_simulations)
simulation_matches['result'] = simulation_matches[['Croatia','France']].apply(
    lambda x: str(x[0]) + " - " + str(x[1]), axis = 1)

frequency_results = pd.DataFrame(simulation_matches['result'].value_counts()).reset_index()
frequency_results.columns = ['result','count']

fig = px.bar(frequency_results, x = frequency_results['result'], y = frequency_results['count'], 
             color = frequency_results['count'])
fig.update_layout(title_text = 'Simulation Result. Croatia 2 (1.38 xG) - (1.10 xG) 4 France')
fig.show()
```

* **Win Simulation**

```python
def result(x):
    if int(x[0]) > int(x[4]):
        return 'Croatia win'
    else:
        if int(x[0]) == int(x[4]):
            return 'Draw'
        else:
            return 'France win'

simulation_matches['win'] = simulation_matches['result'].apply(lambda x: result(x))

win_simulation = pd.DataFrame(simulation_matches['win'].value_counts()).reset_index()
win_simulation.columns = ['result', 'count']

fig = px.pie(win_simulation, names = win_simulation['result'], values = win_simulation['count'])
fig.update_layout(title_text = 'Win Result')
fig.show()
```
